```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
from scipy import fftpack

%matplotlib inline
```

```python
# img = io.imread('/Users/ricardobarroslourenco/PycharmProjects/clouds_rdcep/experimental/image_analysis/original_img.png')


img = plt.imread('/Users/ricardobarroslourenco/PycharmProjects/clouds_rdcep/experimental/image_analysis/original_img.png').astype(float)

plt.figure()
plt.imshow(img, plt.cm.gray)
plt.title('Original image')
```

```python
im_fft = fftpack.fft2(img)
```

```python
def plot_spectrum(im_fft):
    from matplotlib.colors import LogNorm
    # A logarithmic colormap
    plt.imshow(np.abs(im_fft), norm=LogNorm(vmin=5))
    plt.colorbar()
```

```python
plt.figure()
plot_spectrum(im_fft)
plt.title('Fourier transform')
```

```python
img = plt.imread('/Users/ricardobarroslourenco/PycharmProjects/clouds_rdcep/experimental/image_analysis/ae_img.png').astype(float)

plt.figure()
plt.imshow(img, plt.cm.gray)
plt.title('Autoencoded image')

im_fft = fftpack.fft2(img)

plt.figure()
plot_spectrum(im_fft)
plt.title('Fourier transform')
```
